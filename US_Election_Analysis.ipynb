# US Election Analysis

There is corrrelation and possibly a causation between education level and political choice. This analysis is about US 2020 Presidential Election results and catching a possible correlation. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Import election data and check election data.

```python
mydata = pd.read_csv("use.csv", sep=";")

mydata["Bachelors"]=mydata["Bachelors"]/100
mydata["Advanced"]=mydata["Advanced"]/100
mydata.head()
```

## 1. Analysis of educationlevel of States:

States listed by Bachelors or more education.

```python
mydata.sort_values(by=['Bachelors', 'Advanced'])
```

### Observation 1: 13 of 15 least percentage of population with Bachelors or more education voted for Trump.

```python
mydata.sort_values(by=['Bachelors', 'Advanced'], ascending=False)
```

### Observation 2: 15 of 15 highest percentage of population with Bachelors or more education voted for Biden.

## 2. Correlation between Education Level and Choice of President

Bachelors+:

```python
ax = sns.boxplot(x='DR', y='Bachelors', data=mydata);
```

### Observation 3: Apperantly there is a strong correlation between higher Bachelors education and electoral choice. The states that mostly chose Biden have more than double of Bachelors+ educated people

Advanced+

```python
sns.boxplot(x='DR', y='Advanced', data=mydata);
```

## 3. Conclusion

As a conclusion we can easily say that there is a correlation between Bachelors+ education level of population in a state and choice of president of the US in 2020.
